```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Image, display

# Read the image in grayscale
image = cv2.imread('Lenna.png', cv2.IMREAD_GRAYSCALE)

def display_image(image, title):
    plt.figure(figsize=(6, 6))
    plt.imshow(image, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
display_image(image, 'Original Image')
```

# Negative Image

```python
negative_image = 255 - image
display_image(negative_image, 'Image Negative')
```

# Log Transformation

```python
c = 1
log_transformed_image = c * np.log1p(image.astype(np.float64))
log_transformed_image = np.clip(log_transformed_image, 0, 255).astype(np.uint8)
cv2.imwrite("log_image.jpeg", log_transformed_image)
display_image(log_transformed_image, f'Log Transformation (c={c})')
```

```python
c = 10
log_transformed_image = c * np.log1p(image.astype(np.float64))
log_transformed_image = np.clip(log_transformed_image, 0, 255).astype(np.uint8)
cv2.imwrite("log_image.jpeg", log_transformed_image)
display_image(log_transformed_image, f'Log Transformation (c={c})')
```

```python
c = 100
logTransformedImage = c * np.log(1 + image.astype(np.float64))
logTransformedImage = np.array(logTransformedImage, dtype=np.uint8)
cv2.imwrite("log_image.jpeg", logTransformedImage)
display_image(logTransformedImage, f'Log Transformation (c={c})')
```

```python
c = 500
logTransformedImage = c * np.log(1 + image.astype(np.float64))
logTransformedImage = np.array(logTransformedImage, dtype=np.uint8)
cv2.imwrite("log_image.jpeg", logTransformedImage)
display_image(logTransformedImage, f'Log Transformation (c={c})')
```

# Power law transformation

```python
gamma_values = [0.4, 1.0, 2.0]
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for i, gamma in enumerate(gamma_values):
    gamma_image = np.power(image / 255.0, gamma) * 255
    gamma_image = np.clip(gamma_image, 0, 255).astype(np.uint8)
    axes[i].imshow(gamma_image, cmap='gray')
    axes[i].set_title(f'Power-Law Transformation (Gamma={gamma})')
    axes[i].axis('off')
plt.show()
```

# Contrast Stretching

```python
min_val = np.min(image)
max_val = np.max(image)
contrast_stretch = ((image - min_val) / (max_val - min_val) * 255).astype(np.uint8)
display_image(contrast_stretch, 'Contrast Stretching')
```

# Gray level slicing

```python
lower_bound, upper_bound = 100, 200
sliced_image = np.zeros_like(image)
sliced_image[(image >= lower_bound) & (image <= upper_bound)] = 255
display_image(sliced_image, f'Gray Level Slicing ({lower_bound} - {upper_bound})')
```
