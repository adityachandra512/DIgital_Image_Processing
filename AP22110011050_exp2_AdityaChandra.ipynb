# Question 1

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Create Image A (horizontal lines)

```python
image_A = np.zeros((1024, 1024), dtype=np.uint8)
image_A[np.arange(0, 1024, 256)[:, None] + np.arange(128), :] = 127
```

# Create Image B (vertical lines)

```python
image_B = np.zeros((1024, 1024), dtype=np.uint8)
image_B[:, np.arange(0, 1024, 256)[:, None] + np.arange(128)] = 127
```

# Display Image A and B

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Image A (Horizontal Lines)")
plt.imshow(image_A, cmap='gray')
```

```python
plt.subplot(1, 2, 2)
plt.title("Image B (Vertical Lines)")
plt.imshow(image_B, cmap='gray')
plt.show()
```

# a. Image addition of A and B (pixel-wise addition)

```python
image_add = np.clip(image_A.astype(np.int16) + image_B.astype(np.int16), 0, 255).astype(np.uint8)
```

# b. Subtraction of A and B  (Pixel wise subtraction)

```python
image_sub = np.clip(image_A.astype(np.int16) - image_B.astype(np.int16), 0, 255).astype(np.uint8)
```

# c. Multiplying Images of A and B (Pixel-wise multiplication)

```python
image_mul = np.clip(image_A.astype(np.int16) * image_B.astype(np.int16), 0, 255).astype(np.uint8)
```

# Display results of A

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.title("Addition (A + B)")
plt.imshow(image_add, cmap='gray')
```

# Display result of B

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 2)
plt.title("Subtraction (A - B)")
plt.imshow(image_sub, cmap='gray')
```

# Display result of C

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 3)
plt.title("Multiplication (A * B)")
plt.imshow(image_mul, cmap='gray')
plt.show()
```

# Question 2

# a. Grayscale image with sinusoidal intensity

```python
height, width = 256, 1024
x = np.arange(width)
intensity = (127 + 127 * np.sin(2 * np.pi * x / width)).astype(np.uint8)  # Sinusoidal intensity
image_sin = np.tile(intensity, (height, 1))
```

# Display grayscale image

```python
plt.figure(figsize=(10, 5))
plt.title("Grayscale Image (Sinusoidal Intensity)")
plt.imshow(image_sin, cmap='gray', aspect='auto')  # aspect='auto' prevents distortion
plt.colorbar(label="Pixel Intensity")  # Adds a color scale
plt.xlabel("Width (Pixels)")
plt.ylabel("Height (Pixels)")
plt.show()
```

# b. White image with black box at the center

```python
image_box = np.full((256, 256), 255, dtype=np.uint8)  # White image
box_size = 58
start = (256 - box_size) // 2
image_box[start:start + box_size, start:start + box_size] = 0
```

# Display image with black box

```python
plt.figure(figsize=(5, 5))
plt.title("White Image with Black Box")
plt.imshow(image_box, cmap='gray', aspect='auto')
plt.colorbar(label="Pixel Intensity")  # Adds a color scale
plt.xlabel("Width (Pixels)")
plt.ylabel("Height (Pixels)")
plt.show()
```
