# **b. Extract and Display the Color Planes (Red, Green, Blue)**

```python
blue, green, red = cv2.split(image)
```

```python
blue_only_image = np.zeros_like(image)
blue_only_image[:, :, 0] = image[:, :, 0]
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(blue_only_image, cv2.COLOR_BGR2RGB))
plt.title('Blue-Only Image')
plt.axis('off')
plt.show()
```

```python
red_only_image = np.zeros_like(image)
red_only_image[:, :, 2] = image[:, :, 2]
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(red_only_image, cv2.COLOR_BGR2RGB))
plt.title('Red-Only Image')
plt.axis('off')
plt.show()
```

```python
green_only_image = np.zeros_like(image)
green_only_image[:, :, 1] = image[:, :, 1]
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(green_only_image, cv2.COLOR_BGR2RGB))
plt.title('Green-Only Image')
plt.axis('off')
plt.show()
```

# **c. Convert Color Image to Grayscale and Binary**

Conveting the image in Grayscale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
print("Grayscale Image:")
cv2_imshow(gray_image)
```

Conveting this in binary image

```python
_, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)
print("Binary Image:")
cv2_imshow(binary_image)
```

# **d. Resize the Image to Half and Quarter**

conveting this in half size image

```python
half_size_image = cv2.resize(image, (image.shape[1] // 2, image.shape[0] // 2))
print("Half-Size Image:")
cv2_imshow(half_size_image)
```

Conveting this in quarter size image

```python
quarter_size_image = cv2.resize(image, (image.shape[1] // 4, image.shape[0] // 4))
print("Quarter-Size Image:")
cv2_imshow(quarter_size_image)
```

Rotating the image by 90 degree

```python
rotate_90 = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
print("Rotated by 90 Degrees:")
cv2_imshow(rotate_90)
```

Rotataing the image by 180 degree

```python
rotate_180 = cv2.rotate(image, cv2.ROTATE_180)
print("Rotated by 180 Degrees:")
cv2_imshow(rotate_180)
```

Rotataing the image by 45 degree

```python
(h, w) = image.shape[:2]
center = (w // 2, h // 2)
rotation_matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated_45 = cv2.warpAffine(image, rotation_matrix, (w, h))
print("Rotated by 45 Degrees:")
cv2_imshow(rotated_45)
```
