d. Max filtering

e. Min filtering

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('OIP.jpg', cv2.IMREAD_GRAYSCALE)
def display_image(image, title):
    plt.figure(figsize=(6, 6))
    plt.imshow(image, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
display_image(image, 'Original Image')
```

# Using Averaging Filter

```python
window_size = [3, 5, 7, 9]
fig, axes = plt.subplots(1, len(window_size) + 1, figsize=(15, 5))
axes[0].imshow(image, cmap='gray')
axes[0].set_title("Original Image")
axes[0].axis('off')
for i, k in enumerate(window_size):
    window = np.ones((k, k), np.float32) / (k * k)
    blurred_image = cv2.filter2D(image, -1, window)
    axes[i + 1].imshow(blurred_image, cmap='gray')
    axes[i + 1].set_title(f'Average Filter ({k}x{k})')
    axes[i + 1].axis('off')
plt.show()
```

# Using Weight Filter

```python
window_size = [3, 5, 7, 9]
fig, axes = plt.subplots(1, len(window_size) + 1, figsize=(15, 5))
axes[0].imshow(image, cmap='gray')
axes[0].set_title("Original Image")
axes[0].axis('off')
for i, k in enumerate(window_size):
    weighted_blurred = cv2.GaussianBlur(image, (k, k), sigmaX=0)
    axes[i + 1].imshow(weighted_blurred, cmap='gray')
    axes[i + 1].set_title(f'Weighted Filter ({k}x{k})')
    axes[i + 1].axis('off')
plt.show()
```

# Applying Median Filter

```python
window_size = [3, 5, 7, 9]
fig, axes = plt.subplots(1, len(window_size) + 1, figsize=(15, 5))
axes[0].imshow(image, cmap='gray')
axes[0].set_title("Original Image")
axes[0].axis('off')
for i, k in enumerate(window_size):
    median_blur = cv2.medianBlur(image, k)
    axes[i + 1].imshow(median_blur, cmap='gray')
    axes[i + 1].set_title(f'Median Filter ({k}x{k})')
    axes[i + 1].axis('off')
plt.show()
```

# Applying Max Filter

```python
window_sizes = [3, 5, 7, 9]
filtered_images = []
for size in window_sizes:
    kernel = np.ones((size, size), np.uint8)
    max_filtered_image = cv2.dilate(image, kernel)
    filtered_images.append(max_filtered_image)
fig, axes = plt.subplots(1, 4, figsize=(15, 5))
for i, size in enumerate(window_sizes):
    axes[i].imshow(filtered_images[i], cmap='gray')
    axes[i].set_title(f"Max Filter {size}x{size}")
    axes[i].axis('off')
plt.show()
```

# Applying Min Filter

```python
window_sizes = [3, 5, 7, 9]
filtered_images = []
for size in window_sizes:
    window = np.ones((size, size), np.uint8)
    min_filtered_image = cv2.erode(image, window)
    filtered_images.append(min_filtered_image)
fig, axes = plt.subplots(1, 4, figsize=(15, 5))
for i, size in enumerate(window_sizes):
    axes[i].imshow(filtered_images[i], cmap='gray')
    axes[i].set_title(f"Min Filter {size}x{size}")
    axes[i].axis('off')
plt.show()
```
